# Pickle
pickle이란 텍스트 상태의 데이터가 아니라 파이썬 객체 자체를 바이너리 파일로 저장하는 것

즉, 리스트와 클래스 같은 일반적인 파일 입출력 방식의 텍스트 형태 파일로 저장하는 것이 아닌, 그 외 자료형인 dictionary, list, tuple과 같은 형태로 필요한 부분을 핵심적으로 저장하는 것. 그러나 문자열이 아닌 객체를 파일에 쓸 수 없기 때문에 pickle 모듈을 사용해 그 객체 자체를 바이너리로 저장하는 것

따라서 pickle은 파싱 과정이 대폭 생략되어 훨씬 빠르게 처리할 수 있도록 도움

그러나, [취약성](https://watchout31337.tistory.com/167)도 존재하니 유의해야함

## How to use
``` python
pickle.dump('객체','파일) # pickle 파일 저장
pickle.load('파일) # pickle 파일 로드
```

``` python
import pickle

list = ['a', 'b', 'c']

## Save pickle
with open('list.txt', 'wb') as f:
    pickle.dump(list, f)

## Load pickle
with open("data.pickle","rb") as fr:
    data = pickle.load(fr)
print(data)

#['a', 'b', 'c']
```

````python
import pickle
import os
from xml.etree.ElementTree import parse, Element, dump, ElementTree

folder_path = "/content/korean-speech-recognition-quartznet/datasets/meta/aihub/"

with open("test.txt", 'wb') as f:
    for dirname, _, filenames in os.walk(folder_path):
        # print(dirname) #/content/korean-speech-recognition-quartznet/datasets/meta/aihub/KsponSpeech_0027
        # print(filenames) #[KsponSpeech_023872.txt, KsponSpeech_023873.txt, ]
        subdirname = dirname.replace(folder_path, "")
        for filename in sorted(filenames):
            # print(filename) # KsponSpeech_023872.txt
            if filename.endswith(".txt"):
                try:
                    with open(subdirname+"/"+filename, 'rb') as txtfile:
                        data = txtfile.readlines() 
                        # print(data) # [b'\xbc\xd2\xbc\xb3 \xc0\xd0\xbe\xfa\xb4\xc2\xb5\xa5? b/\r\n']
                        char2idx = dict()
                        char2idx = {char:idx for idx, char in enumerate(data)}
                        idx2char = list()
                        idx2char = {idx:char for idx, char in enumerate(data)}
                        print(char2idx)
                        pickle.dump(data, f)
                        txtfile.close()
                except:
                    print(filename+': no such file')
    f.close()

with open("test.txt", 'rb') as f:
    data = pickle.load(f)
    print(data)
    f.close()
````
